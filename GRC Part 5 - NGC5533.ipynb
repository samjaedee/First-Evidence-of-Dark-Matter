# Part 5 - Rotation Curve of The Galaxy NGC5533

In this notebook, we plot the rotation curve of NGC5533. We then inspect in detail the velocity of each luminous components (bulge, disk, and gas) and calculate the total velocity of only the luminous components. We notice that there are some discrepancies between the observed velocity and the total velocity of the luminous components. We then add the velocity of dark matter component and find that it perfectly matches the observed velocity. This proves the existence of dark matter!

### Part A: Plot The Observed Velocity and The Velocity of The Luminous Components

```python
# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
# Load the data of the galaxy
r_obs, v_obs, verr_obs = np.genfromtxt('noord-100kpc-observed.txt', usecols = (0, 1, 3), unpack = True)
r_bulge, v_bulge       = np.genfromtxt('noord-120kpc-bulge.txt', skip_header = 1, unpack = True)
r_disk, v_disk         = np.genfromtxt('noord-120kpc-disk.txt', skip_header = 1, unpack = True)
r_gas, v_gas           = np.genfromtxt('noord-120kpc-gas.txt', skip_header = 1, unpack = True)
r_bh, v_bh             = np.genfromtxt('noord-120kpc-blackhole.txt', skip_header = 1, unpack = True)
```

```python
# Create a plot 
r_lowerlim = min(r_obs) + 0.01
r_upperlim = max(r_obs)
r = np.linspace(r_lowerlim,r_upperlim,125)

plt.figure(figsize=(14,8),dpi=100)
plt.title('Rotation Curve of NGC5533 - Luminous Matter and Blackhole', fontsize=14)
plt.xlabel('Radius (kpc)', fontsize=12)
plt.ylabel('Velocity (km/s)', fontsize=12)
plt.xlim(0,r_upperlim)
plt.ylim(0,350) 

f_disk = make_interp_spline(r_disk, v_disk)
f_gas = make_interp_spline(r_gas, v_gas)
f_bulge = make_interp_spline(r_bulge, v_bulge)
f_bh = make_interp_spline(r_bh, v_bh)

plt.errorbar(r_obs, v_obs, yerr=verr_obs, marker = 'o', markersize=8, linestyle='none', \
             label="Observed velocity", color="grey")
plt.plot(r, f_disk(r),  label="Disk velocity",  color="royalblue")
plt.plot(r, f_gas(r),   label="Gas velocity",   color="gold")
plt.plot(r, f_bulge(r), label="Bulge velocity", color="seagreen")
plt.plot(r, f_bh(r),    label="Blackhole",      color='maroon')

plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.show()
```

### Part B: Add Dark Matter!

Data for NGC5533
* Central mass density = 0.31e9 (in solar mass/kpc^3), Source: Noordermeer (2007)
* Core radius = 1.4 (in kpc), Source: Noordermeer (2007)
* Mass of central black hole in NGC5533 = 2.7e9 (in solar masses), Source: Noordermeer (2007)

```python
# Set parameters
rho0 = 0.31e9          # central mass density (in solar mass/kpc^3)
rc = 1.4               # core radius (in kpc)
radius = 10            # radius (in kpc)
G = 4.300e-6           # Gravitational constant (kpc/solar mass*(km/s)^2)

# Display the two dark matter parameters
print("Dark matter halo parameters: ")
print("     Central mass density: {:.2e} solar mass/kpc^3".format(rho0))
print("     Core radius: {:.2f} kpc".format(rc))

# Define the function that will calculate the dark matter halo velocity
def halo_v(r,rho0,rc):
    v = np.sqrt(4*np.pi*G*rho0*rc**2*(1 - rc/r * np.arctan(r/rc)))
    return v

# Display the velocity at the selected radius
print("     Velocity at {} kpc: {:.3f} km/s".format(radius,halo_v(radius,rho0,rc)))
```

```python
# Create the plot of all the components of the galaxy
r_lowerlim = min(r_obs) + 0.01
r_upperlim = max(r_obs)
r = np.linspace(r_lowerlim,r_upperlim,125)

plt.figure(figsize=(14,8),dpi=100)
plt.title('Rotation Curve of NGC5533 - Luminous Matter, Blackhole, and Dark Matter', fontsize=14)
plt.xlabel('Radius (kpc)', fontsize=12)
plt.ylabel('Velocity (km/s)', fontsize=12)
plt.xlim(0,r_upperlim)
plt.ylim(0,350) 

plt.errorbar(r_obs, v_obs, yerr=verr_obs, marker = 'o', markersize=8, linestyle='none', \
             label="Observed velocity", color="grey")
plt.plot(r, halo_v(r,rho0,rc),   label='Dark matter halo',  color='y', )  
plt.plot(r, f_disk(r),           label="Disk velocity",    color="royalblue")
plt.plot(r, f_gas(r),            label="Gas velocity",     color="gold")
plt.plot(r, f_bulge(r),          label="Bulge velocity",   color="seagreen")
plt.plot(r, f_bh(r),             label="Blackhole",        color='maroon')

plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.show()
```

### Part C: Plot The Total Velocity With Dark Matter and The Total Velocity Without Dark Matter

```python
#Set paramaters
massbh = 2.7e9         # mass of the central black hole (in solar masses)

# Scaling parameters
bpref = 0.75           # prefactor for bulge: 0.75 or 1
dpref = 1              # prefactor for disk: 0.75 or 1
gpref = 1              # prefactor for gas
bhpref = 1             # prefactor for black hole

# Total velocity WITHOUT dark matter halo component but with adjustable prefactors
def total_v_noDM(r):
    v = np.sqrt( (bpref * f_bulge(r))**2                   # bulge component + scaling parameter
               + (dpref * f_disk(r))**2                    # disk component + scaling parameter
               + (gpref * f_gas(r))**2                     # gas component + scaling parameter
               + (bhpref * f_bh(r))**2)                    # central supermassive black hole component + scaling factor
    return v

# Total velocity WITH dark matter halo component but with adjustable prefactors
def total_v_DM(r, rho0, rc):
    v = np.sqrt( 
        (bpref * f_bulge(r))**2                            # bulge component + scaling parameter
        + (dpref * f_disk(r))**2                           # disk component + scaling parameter
        + (gpref * f_gas(r))**2                            # gas component + scaling parameter
        + (bhpref * f_bh(r))**2                            # central supermassive black hole component + scaling factor
        + (halo_v(r,rho0,rc))**2)                          # dark matter halo component
    return v
```

```python
# Create the plot of the rotation curve of the galaxy
r_lowerlim = 0.01
r_upperlim = max(r_obs)
r = np.linspace(r_lowerlim,r_upperlim,100)

plt.figure(figsize=(14,8),dpi=100)
plt.title('Rotation Curve of NGC5533', fontsize=14)
plt.xlabel('Radius (kpc)', fontsize=12)
plt.ylabel('Velocity (km/s)', fontsize=12)
plt.xlim(0,r_upperlim)
plt.ylim(0,350) 

plt.errorbar(r_obs, v_obs, yerr=verr_obs, marker = 'o', markersize=8, linestyle='none', \
             label="Observed velocity", color="grey")
plt.plot(r, halo_v(r,rho0,rc),     label='Dark matter halo',        color='y')  
plt.plot(r, f_disk(r),             label="Disk velocity",           color="royalblue")
plt.plot(r, f_gas(r),              label="Gas velocity",            color="gold")
plt.plot(r, f_bulge(r),            label="Bulge velocity",          color="seagreen")
plt.plot(r, f_bh(r),               label="Blackhole",               color='maroon')
plt.plot(r, total_v_DM(r,rho0,rc), label='Total curve with DM',     color='red')
plt.plot(r, total_v_noDM(r),       label='Total curve without DM',  color='darkorange', )

plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.savefig('RotationNGC5533.png')
```
